```python
import pandas as pd
```

#### Read from original dataset
<a href="https://www.kaggle.com/datasets/esraamadi/saudi-arabia-weather-history">Saudi Arabia Weather History</a>

```python
# Read csv file (weather-sa-2017-2019-clean.csv)
df = pd.read_csv('../Dataset/weather-sa-2017-2019-clean.csv', index_col=0)
df
```

```python
# Add new column to combine the (date) and (time) columns as a string value
df['string_timestamp'] = df['date'] + ' ' +  df['time']
df['string_timestamp']
```

```python
df['timestamp'] = pd.to_datetime(df['string_timestamp'], format='%d %B %Y %H:%M')
```

```python
# Generate the new ID in the format 'city-year-month-day-time'
df['ID'] = df['city'] + '-' + df['year'].astype(str) + '-' + df['month'].astype(str) + '-' + df['day'].astype(str) + '-' + df['time']
```

```python
new_weather_df = df[['ID', 'city', 'timestamp', 'weather', 'temp', 'wind', 'humidity', 'barometer', 'visibility']]
new_weather_df
```

```python
def dataset_by_city(*cities):
    
    for city in cities:
        if " " in city:
            city_with_underscore = city.replace(" ", "_")
            city = city_with_underscore
        city_weather = new_weather_df.loc[new_weather_df['city'] == f"{city}"]
        city_weather.to_csv(f"../Reorganize-dataset/{city}_weather_log.csv", index=False)
```

```python
dataset_by_city('Jawf', 'Baha', 'Najran', 'Northern boarder', 'Jazan', 'Assir', 
                'Tabuk', 'Mecca', 'Riyadh', 'Madina', 'Hail', 'Qassim')
```
